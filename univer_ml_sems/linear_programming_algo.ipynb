```python
import numpy as np

def remove_dominated_rows_cols(array):
    rows_to_keep = list(range(array.shape[0]))
    cols_to_keep = list(range(array.shape[1]))

    # Удаляем доминируемые строки
    dominated_rows = set()
    for i in range(len(rows_to_keep)):
        for j in range(len(rows_to_keep)):
            if i != j:
                row_i = array[rows_to_keep[i], :]
                row_j = array[rows_to_keep[j], :]
                # Строка i доминируется строкой j, если все элементы row_i <= row_j
                # и хотя бы в одном элементе строго меньше
                if all(row_i <= row_j) and any(row_i < row_j):
                    dominated_rows.add(rows_to_keep[i])

    rows_to_keep = [i for i in rows_to_keep if i not in dominated_rows]

    # Удаляем доминирующие столбцы
    dominated_cols = set()
    for i in range(len(cols_to_keep)):
        for j in range(len(cols_to_keep)):
            if i != j:
                col_i = array[:, cols_to_keep[i]]
                col_j = array[:, cols_to_keep[j]]
                # Столбец j доминируется столбцом i, если во всех строках col_i >= col_j
                # и хотя бы в одном строго больше
                if all(col_i >= col_j) and any(col_i > col_j):
                    dominated_cols.add(cols_to_keep[j])

    cols_to_keep = [j for j in cols_to_keep if j not in dominated_cols]

    return array[np.ix_(rows_to_keep, cols_to_keep)]


alpha, beta, n, k = 0.2, 0.3, 20, 4
array = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == j:
            array[i][j] = 0
        elif abs(i - j) <= k:
            array[i][j] = i - j
        elif abs(i - j) > k:
            array[i][j] = alpha * (i + 1) - beta * (j + 1)

for i in range(n):
    for j in range(n):
      array[i][j] += 7

filtered_array = remove_dominated_rows_cols(array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)
filtered_array = remove_dominated_rows_cols(filtered_array)

print("Изначальная матрица")
for row in array:
    print(" ".join(f"{num:6.1f}" for num in row))

print("\n\nМатрица только с доминирующими строками и столбцами")
for row in filtered_array:
    print(" ".join(f"{num:6.1f}" for num in row))

print(filtered_array.shape)

wb = Workbook()
ws = wb.active

# Записываем данные массива
for i, row in enumerate(filtered_array, start=1):
    for j, value in enumerate(row, start=1):
        ws.cell(row=i, column=j, value=value)

# Сохраняем файл
wb.save("filtered_array_1_min.xlsx")

for i, row in enumerate(array, start=1):
    for j, value in enumerate(row, start=1):
        ws.cell(row=i, column=j, value=value)

wb.save("filtered_array_1_max.xlsx")
```
