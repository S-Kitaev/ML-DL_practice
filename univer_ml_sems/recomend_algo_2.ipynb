# Напишу функицию насчета косинусного расстояния между двумя векторами

```python
def cosinus(list1, list2):
  return round((np.dot(list1, list2))/(np.linalg.norm(list1) * np.linalg.norm(list2)),3)
```

```python
print(f"Косинусное расстояние между пользователями User i и User j: {cosinus(np.array(df.loc['User i']), np.array(df.loc['User j']))}")
print(f"Косинусное расстояние между пользователями User i и User k: {cosinus(np.array(df.loc['User i']), np.array(df.loc['User k']))}")
print(f"Косинусное расстояние между пользователями User k и User j: {cosinus(np.array(df.loc['User k']), np.array(df.loc['User j']))}")
```
