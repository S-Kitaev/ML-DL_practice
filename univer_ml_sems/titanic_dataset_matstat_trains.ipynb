```python
# 6. Какое самое популярное женское имя было на корабле? Извлеките из полного имени пассажира (колонка Name)
# его личное имя

# сначала определим всех женщин и сделаем series с их именами, далее будем данный список аккуратно чистить, пока не
# останутся только имена женщин

data_fem = data[data["Sex"] == "female"]
data_fem = data_fem["Name"]
# print(data_fem.head(20))

# Заметим, что имя всегда пишется после слова mrs или miss, проверим, что данная запись есть у всех женщин
data_fem.head(25)
```

```python
import re

print("Всего женщин:", len(data_fem))
pattern = re.compile(r'\b(miss\.|mrs\.)', re.IGNORECASE)

def fem_filter(name):
    return bool(pattern.search(name))

count = data_fem.apply(fem_filter).sum()

print(f"Проверка соответствия количества: {count}")
```

```python
# Отдельно запишем эти 7 имен, и оставшиеся имена в два датафрейма

data_fem_nomiss = data_fem[~data_fem.apply(fem_filter)]
data_fem_miss = data_fem[data_fem.apply(fem_filter)]

# Создаем DataFrame с отфильтрованными именами
data_fem_nomiss
```

```python
# тут можно вручную пройтись и все имена записать. Вторая строчка вообще не относится к женщинам

data_fem_nomiss = pd.Series(['Pauline','Reynaldo', 'Christiana','Emma','Berthe','Lucy','Alice'])
```

```python
# для остальных оставим только запись после miss. mrs.

pattern = re.compile(r'.*\b(?:miss\.|mrs\.)\s*', re.IGNORECASE)

def remove_female_title(name):
    return pattern.sub("", name)

# Применяем функцию ко всем записям
data_fem_miss = data_fem_miss.apply(remove_female_title)
data_fem_miss.head(5)
```

```python
# осталось всего 2 вида имен - такие, где есть скобки, и такие, где их нет, разделим их

pattern = re.compile(r'\(.*?\)')

data_fem_sk = data_fem_miss[data_fem_miss.apply(fem_filter)]
data_fem_nosk = data_fem_miss[~data_fem_miss.apply(fem_filter)]
data_fem_nosk.head(5)
```

```python
# Там, где нет скобок возьмем первое слово - это имя

data_fem_nosk = data_fem_nosk.str.split().str[0]
data_fem_nosk.head(5)
```

```python
# А там, где есть скобки возьмем первое слово после скобок

pattern = re.compile(r'\((\w+)')

def fem_with_sk(name):
    match = pattern.search(name)
    return match.group(1) if match else None

data_fem_sk = data_fem_sk.apply(fem_with_sk)
data_fem_sk.head(5)
```

```python
# соединим все 3 получившихся series и посчитаем самое популярное имя

fem_names = pd.concat([data_fem_sk,data_fem_nosk,data_fem_nomiss])

fem_names.value_counts().head(10)
```
