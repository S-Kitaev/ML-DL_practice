# Линейная регрессия - задача 2

```python
points_auto = [[6,13],[7,16],[8,15],[9,20],[10,19],[11,21]]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# напишу единую функцию, которая будет принимать массив данных в формате [[x,y]] и выдавать коэфициенты a и b и выборочный коэфициент корреляции r

def regression_calculator(points):

  X = [point[0] for point in points]
  Y = [point[1] for point in points]

  xy = []
  for elem in range(len(points)):
    xy.append(X[elem]*Y[elem])
  sum_xy = sum(xy)

  sum_x = sum(X)

  sum_y = sum(Y)

  x_2 = [x*x for x in X]
  sum_x_2 = sum(x_2)

  y_2 = [y*y for y in Y]
  sum_y_2 = sum(y_2)


  n = len(points)

  a = (sum_xy - sum_y * sum_x / n) / (sum_x_2 - sum_x**2 / n)

  b = (sum_y - a * sum_x) / n

  r = (n * sum_xy - sum_x * sum_y) / ((n * sum_x_2 - sum_x**2)**0.5 * (n * sum_y_2 - sum_y**2)**0.5)

  return a, b, r
```

```python
# Напишу функцию визуализации данных. Она принимает массив данных и выдает картинку c данными коэфициента корреляции и получившейся прямой

def vizual_regression(points):
    a, b, r = regression_calculator(points)

    X = [point[0] for point in points]
    Y = [point[1] for point in points]

    x_min, x_max = min(X), max(X)
    x_line = np.linspace(x_min, x_max, 100)
    y_line = a * x_line + b

    plt.figure(figsize=(8, 6))
    plt.scatter(X, Y, color='blue', label='Точки')
    plt.plot(x_line, y_line, color='red', label=f'Линия регрессии (y={round(a,2)}x+{round(b,2)})')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Линейная регрессия (Коэффициент корреляции r={round(r,2)})')
    plt.legend()
    plt.grid()
    plt.show()
```
