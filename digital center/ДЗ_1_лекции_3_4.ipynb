Домашнее задание к лекции №3

Загрузить список из 1000 вакансий программиста для Москвы и Санкт-Петербурга с api HH. Вакансия должна содержать id вакансии, название, зарплату от, зарплату до, валюту зарплаты, включение налога, город, название компании, обязанности и требования.

Используя список вакансии в обязанностях выявить список из 5 наиболее частых слов, и с помощью множеств оставить только те вакансии, которые их содержат
Работа должна быть сделана в Google Colab, указать ссылку на блокнот Colab.
Результат работы кода должен быть продемонстрирован путем запуска ячейки в выходных данных.


```
import requests

salary = lambda x: [x['salary']['from'], x['salary']['to'], x['salary']['currency'], x['salary']['gross']] if x['salary'] is not None else [None]*4
snippet = lambda x: [x['snippet']['responsibility'], x['snippet']['requirement']] if x['snippet'] is not None else [None]*2

vac = []

for i in range(10):
  url = 'https://api.hh.ru/vacancies?text=Программист&per_page=100&area=1&area=2&page={}'.format(str(i))
  vac += [[x['id'],x['name']]+ salary(x) + [x['employer']['name']] + [x['area']['name']] + snippet(x) for x in requests.get(url).json()['items']]

s = []
treb = []
for i in range(len(vac)):   # делаю 1 список со всеми словами обязанностей
  resp = vac[i][-2]
  if resp != None:
    resp = resp.split()
    treb += resp

for i in range(len(treb)):
  treb[i] = treb[i].lower()
  if treb[i][-1] in '.,()-' :  # убираю лишние знаки препинания в некоторых словах
    treb[i] = treb[i][:-1]
  elif treb[i][0] == '(':
    treb[i] = treb[1:]

dop = []
for i in range(len(treb)-116):  # проверяю список на наличие вложенных списков и создаю дополнительный список куда добавляю элементы вложенного списка
  if type(treb[i]) == list:
    a = treb.pop(i)
    dop.extend(a)

treb = treb + dop # объединяю изначальный и дополнительный списки

ptreb = []
for i in treb:
  if type(i) != list: # проверяю новый список на наличие вложенных списков и создаю новый список куда добавляю только строки
    ptreb.append(i)

for i in range(len(ptreb)):
  ptreb[i] = ptreb[i].lower() # перевожу слова в нижний регистр

c = 0
for i in range(len(ptreb)):   # перепроверяю еще раз что в списке только строки
  if type(ptreb[i]) != str:
    c += 1

r = {}
for i in ptreb: # словарь, в котором подсчитано количество вхождений каждого слова
  r[i] = r.get(i,0) + 1

r_sorted = sorted(r.items(),key=lambda x : x[1],reverse=True) # сортировка словаря
r5 = r_sorted[:5] # срез словаря с 5ю самыми частыми словами

slov5 = []  # для удобства делаю дополнительный список с 5ю словами
for i in range(5):
  slov5.append(r5[i][0])

# фильтрую список и получаю ответ
answer = list(filter(lambda x : x if x[-2] != None and slov5[0] in x[-2].lower() and slov5[1] in x[-2].lower() and slov5[2] in x[-2].lower() and slov5[3] in x[-2].lower() and slov5[4] in x[-2].lower() else False,vac))

# вывожу множество подходящих id
for i in range(len(answer)):
  answer[i] = answer[i][0]

#ответ
print(set(answer))
```

Создать пример для демонстрации работы программы по заказу пиццы.
Работа должна быть сделана в Google Colab, указать ссылку на блокнот Colab.
Результат работы кода должен быть продемонстрирован путем запуска ячейки в выходных данных.


```
class Pizza:
  def __init__(self):
    self.name = str
    self.dough = str
    self.topping = list
    self.cost = float
  def cook(self):
    print('The pizza is cooking. Please, waiting!')
  def bake(self):
    print('The pizza is baking. Please, waiting!')
  def cut(self):
    print("We're already cutting the pizza. We'll serve it on the table soon!")
  def wrap(self):
    print('We wrapped the pizza. You will get it soon!')

class PizzaPepperoni(Pizza):
  def __init__(self):
    self.name = 'Pepperoni'
    self.dough = 'thin'
    self.topping = ['pepperoni','cheese','tomato','sause']
    self.cost = 650.0

class PizzaBBQ(Pizza):
  def __init__(self):
    self.name = 'BBQ'
    self.dough = 'thick'
    self.topping = ['mushrooms','bacon','chicken','BBQ - sause']
    self.cost = 700.0

class Order(Pizza):
  def __init__(self):
    self.ordered_pizza = []
    self.Order_ID = int

  def add(self,pizza):
    self.ordered_pizza.append(pizza)

  def cost(self):
    result_cost = 0
    for i in self.ordered_pizza:
      result_cost += i.cost
    return result_cost

  def complete(self):
    print('The pizza is ready!')

# далее - имитация заказа, клиент заказал две разные пиццы - одну пипперони и одну барбекю
# фразы, которые клиент будет видеть на экране в процессе

pizza1 = PizzaPepperoni()
pizza2 = PizzaBBQ()

print('Рады приветствовать в нашей пиццерии! Что будете заказывать?')
print()

order1 = Order()
order1.add(pizza1)
order1.add(pizza2)
order1.Order_ID = 2781528 # число выдумано

print('Прекрасный выбор! Наши гости очень любят эти пиццы, надеемся, вам тоже понравятся!')
print('Ваш заказ №', order1.Order_ID, 'оформлен.', 'Сумма к оплате:', order1.cost(), 'руб.')
print('Ожидайте. Пока ждете, можете следить за тем, на какой стадии приготовления находятся ваши пиццы!')
print()

order1.cook()
order1.bake()
order1.cut()
order1.wrap()
order1.complete()
print()

print('Будем рады видеть вас снова!')
```
